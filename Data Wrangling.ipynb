```python
#Drop data that is more than 8 hours in duration
df = df.loc[df['Duration'] < 28800]
```

```python
#Drop data that is more than 4 std deviations in distance
df = df.loc[df['Distance'] < 40000]
```

```python
#Drop data where stopover > 4 hours
df = df.loc[df['StopoverDuration'] < 14400]
```

```python
df.describe().T
```

```python
#Fix date data
df['Departure']=pd.to_datetime(df.Departure)
```

```python
df['Return']=pd.to_datetime(df.Return)
```

```python
df['Departure']=df['Departure'].dt.strftime('%Y-%m-%d %H:%M')
df['Return']=df['Return'].dt.strftime('%Y-%m-%d %H:%M')
```

```python
df.head()
```

```python
len(df.Weekday.unique())
```

```python
#Create columns for month, day, and weekday
df['Month'] = pd.DatetimeIndex(df['Departure']).month
df['Day'] = pd.DatetimeIndex(df['Departure']).day
df['Weekday'] = pd.DatetimeIndex(df['Departure']).weekday
```

```python
#Create new column for season
doy = pd.DatetimeIndex(df['Departure']).dayofyear
seasons = np.array(['Winter', 'Spring', 'Summer', 'Fall', 'Winter'])
f = np.searchsorted([80, 172, 264, 355], doy)
df['Season'] = seasons[f]
```

```python
#Determine time of day stamp for departure and return times
b = [0,4,8,12,16,20,24]
l = ['Late Night', 'Early Morning','Morning','Noon','Evening','Night']
df['DepartureToD'] = pd.cut(pd.DatetimeIndex(df['Departure']).hour, bins=b, labels=l, include_lowest=True)
df['ReturnToD'] = pd.cut(pd.DatetimeIndex(df['Return']).hour, bins=b, labels=l, include_lowest=True)
```

```python
#Ensure data integrity
df.DepartureToD.value_counts()
```

```python
#Bring in climate data
climate_df = pd.read_excel('climate.xls')
```

```python
#Preview climate data
climate_df.head()
```

```python
#Convert local date column to date time
climate_df['LOCAL_DATE']=pd.to_datetime(climate_df['LOCAL_DATE'])
```

```python
#merge precipitation data to bike rental data
cleandf = pd.merge(df, climate_df,how='left', left_on=[pd.DatetimeIndex(df['Departure']).month, pd.DatetimeIndex(df['Departure']).day, pd.DatetimeIndex(df['Departure']).year], right_on=[pd.DatetimeIndex(climate_df['LOCAL_DATE']).month, pd.DatetimeIndex(climate_df['LOCAL_DATE']).day, pd.DatetimeIndex(climate_df['LOCAL_DATE']).year])
```

```python
#ensure no data was lost
cleandf.info()
```

```python
#drop unneeded columns and rename new ones
cleandf.drop(cleandf.columns[[0,1,2,20,21,22,23]], axis = 1 , inplace = True)
cleandf.rename(columns = {'TOTAL_PRECIPITATION':'Precipitation'},inplace=True)
```
