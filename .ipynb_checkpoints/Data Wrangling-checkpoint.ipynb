```python
#Bin age groups
bins = [0,18, 30, 40, 50, 60, 70, 120]
labels = ['<18','18-30', '30-39', '40-49', '50-59', '60-69', '70+']
rawdata['agerange'] = pd.cut(rawdata.Age, bins, labels = labels,include_lowest = True)
```

```python
rawdata.head()
```

```python
#Replace class dependent survey areas with median
rawdata.groupby("Class")['Seat comfort'].transform(lambda x: x.fillna(np.mean(x), inplace=True))
#rawdata['Food and drink'] = rawdata.groupby("Class")['Food and drink'].transform(lambda x: x.fillna(np.median(x)))
#rawdata['Leg room service'] = rawdata.groupby("Class")['Leg room service'].transform(lambda x: x.fillna(np.median(x)))
#rawdata['Checkin service'] = rawdata.groupby("Class")['Checkin service'].transform(lambda x: x.fillna(np.median(x)))
#rawdata['On-board service'] = rawdata.groupby("Class")['On-board service'].transform(lambda x: x.fillna(np.median(x)))
#rawdata['Baggage handling'] = rawdata.groupby("Class")['Baggage handling'].transform(lambda x: x.fillna(np.median(x)))
```

```python
#Replace age range dependent survey areas with median
rawdata['Online support'] = rawdata.groupby("agerange")['Online support'].transform(lambda x: x.fillna(np.median(x)))
rawdata['Ease of Online booking'] = rawdata.groupby("agerange")['Ease of Online booking'].transform(lambda x: x.fillna(np.median(x)))
rawdata['Online boarding'] = rawdata.groupby("agerange")['Online boarding'].transform(lambda x: x.fillna(np.median(x)))
rawdata['Inflight wifi service'] = rawdata.groupby("agerange")['Inflight wifi service'].transform(lambda x: x.fillna(np.median(x)))
rawdata['Inflight entertainment'] = rawdata.groupby("agerange")['Inflight entertainment'].transform(lambda x: x.fillna(np.median(x)))
```

```python
#Replace type of travel dependent survey areas with median
rawdata['Departure/Arrival time convenient'] = rawdata.groupby("Type of Travel")['Departure/Arrival time convenient'].transform(lambda x: x.fillna(np.median(x)))
rawdata['Gate location'] = rawdata.groupby("Type of Travel")['Gate location'].transform(lambda x: x.fillna(np.median(x)))
```

```python
#Replace non dependent survey areas with median
rawdata['Cleanliness'].fillna(np.median(rawdata['Cleanliness']), inplace = True)
```
